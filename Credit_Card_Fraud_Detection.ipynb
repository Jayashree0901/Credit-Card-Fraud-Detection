```python
df1 = df.drop(columns=['Class'])  # creating a new dataframe eliminating the class column as it is our target feature, so we do not include it in x.
df1.head()
```

```python
x1=df[['V4', 'V11']].values  #based on the positive strongly correlated values
```

Initialising x and y values

```python
x=df1[:].values
y=df['Class'].values
```

Spliting our dataset into training and testing sets for both the feature selections that we have done above

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=42)
x_train1, x_test1, y_train1, y_test1 = train_test_split(x1,y,test_size=0.25, random_state=42)
# x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25, random_state=42)
# x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

Standardizing our features

```python
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
x_test1 = sc.fit_transform(x_test1)
x_train1 = sc.fit_transform(x_train1)
```

```python
x_train
```

# **Random Forest Classifier**

```python
from sklearn.ensemble import RandomForestClassifier
mod = RandomForestClassifier(n_estimators=100, criterion='entropy')
mod.fit(x_train,y_train) # model building
```

```python
from sklearn.ensemble import RandomForestClassifier
mod0 = RandomForestClassifier(n_estimators=100, criterion='entropy')
mod0.fit(x_train1,y_train1) # model building
```

```python
y_pred0 = mod0.predict(x_test1)
```

```python
from sklearn.metrics import accuracy_score, classification_report
print(classification_report(y_test, y_pred))
```

```python
print(accuracy_score(y_test, y_pred))
```
